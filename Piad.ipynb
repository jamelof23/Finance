```
import requests
import pandas as pd
import concurrent.futures
import time

# Your Financial Modeling Prep API key
API_KEY = 'KcsOu8cJesk2GpAJ0DXN4a5Il3DlLUX1'
BASE_URL = 'https://financialmodelingprep.com/api/v4/insider-trading'

# Endpoints to get all symbols from Dow Jones, S&P 500, and NASDAQ
INDEX_URLS = {
    "dow_jones": f"https://financialmodelingprep.com/api/v3/dowjones_constituent?apikey={API_KEY}",
    "sp500": f"https://financialmodelingprep.com/api/v3/sp500_constituent?apikey={API_KEY}",
    "nasdaq": f"https://financialmodelingprep.com/api/v3/nasdaq_constituent?apikey={API_KEY}"
}

def fetch_symbols():
    """Fetch symbols from Dow Jones, S&P 500, and NASDAQ, and remove duplicates."""
    all_symbols = set()
    for index, url in INDEX_URLS.items():
        try:
            response = requests.get(url)
            response.raise_for_status()
            data = response.json()
            symbols = [item['symbol'] for item in data]
            all_symbols.update(symbols)
        except Exception as e:
            print(f"Error fetching {index} symbols: {e}")
    return list(all_symbols)

def fetch_insider_trading(symbol):
    """Fetch insider trading data for a given symbol."""
    url = f"{BASE_URL}?symbol={symbol}&apikey={API_KEY}"
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()
        if data:
            return symbol, data
        else:
            return symbol, None
    except Exception as e:
        return symbol, None

def fetch_all_data(symbols):
    """Fetch insider trading data for all symbols using parallel processing."""
    fetched_data = []
    total_symbols = len(symbols)

    with concurrent.futures.ThreadPoolExecutor(max_workers=10) as executor:
        futures = {executor.submit(fetch_insider_trading, symbol): symbol for symbol in symbols}
        completed = 0

        for future in concurrent.futures.as_completed(futures):
            symbol, data = future.result()
            if data:
                for transaction in data:
                    transaction['Symbol'] = symbol
                    fetched_data.append(transaction)
                print(f"Fetched data for {symbol} successfully.")
            else:
                print(f"No data found for {symbol}.")

            # Update progress
            completed += 1
            print(f"{completed} out of {total_symbols} symbols completed.")

    return fetched_data

def save_to_csv(data, filename='Insider_Transactions_FMP.csv'):
    """Save fetched data to a CSV file."""
    df = pd.DataFrame(data)
    df.to_csv(filename, index=False)
    print(f"\nData saved to {filename}")

# Step 1: Fetch symbols from Dow Jones, S&P 500, and NASDAQ
symbols = fetch_symbols()
print(f"Total unique symbols fetched: {len(symbols)}")

# Step 2: Fetch insider trading data for all symbols and display progress
start_time = time.time()
fetched_data = fetch_all_data(symbols)
end_time = time.time()

print(f"\nFetched insider trading data for {len(fetched_data)} entries in {end_time - start_time:.2f} seconds.")

# Step 3: Save the fetched data to a CSV file on your computer
save_to_csv(fetched_data)
```

```
import requests
import pandas as pd
import time

# Your Financial Modeling Prep API key
API_KEY = 'KcsOu8cJesk2GpAJ0DXN4a5Il3DlLUX1'
BASE_URL = 'https://financialmodelingprep.com/api/v4/insider-trading'

# Read symbols from CSV file
symbols_file = '/content/sample_data/Symbols.csv'
symbols_df = pd.read_csv(symbols_file)
symbols_list = symbols_df['Symbol'].tolist()

# Prepare an empty DataFrame to store the results
all_data = pd.DataFrame()

# Function to fetch insider trading data for a given symbol
def fetch_insider_trading(symbol):
    url = f"{BASE_URL}?symbol={symbol}&apikey={API_KEY}"
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()

        # Check if there are results in the response
        if data:
            return pd.DataFrame(data)
        else:
            print(f"No data found for {symbol}")
            return pd.DataFrame()
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred for {symbol}: {http_err}")
    except Exception as err:
        print(f"Error occurred for {symbol}: {err}")
    return pd.DataFrame()

# Loop through each symbol, fetch data, and append to the DataFrame
for symbol in symbols_list:
    print(f"Fetching insider trading data for {symbol}...")
    df = fetch_insider_trading(symbol)
    if not df.empty:
        df['Symbol'] = symbol  # Add a column for the symbol
        all_data = pd.concat([all_data, df], ignore_index=True)
    # Respect the API rate limit (5 requests per minute for free plan)
    time.sleep(12)

# Save the compiled data to an Excel file
output_file = '/content/Insider_Transactions_FMP.xlsx'
all_data.to_excel(output_file, index=False)

print(f"Data saved to {output_file}")
```
